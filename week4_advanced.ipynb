## [MY CODE] prediction 함수 작성

```python
def prediction(problem, openai_client, gpt_model = "gpt-3.5-turbo"):
    paragraph = problem['paragraph']
    question = problem['question']
    question_plus = problem.get('question_plus', '없음')
    answer = problem['answer']
    choices = problem['choices']
    score = problem['score']

    # 선택지 포맷 변경
    if(isinstance(choices, list)):
        tmp = []
        for i in range(len(choices)):
            tmp.append(f'{i+1}번 - {choices[i]}')

        choices = '\n'.join(tmp)

    # prompt format
    base_prompt = """국어 시험 문제를 푸는 대한민국의 고3 수험생으로서 문제의 정보를 바탕으로 다음 문제의 답을 구하세요.

문제를 풀이할 때, 반드시 지문을 참고하세요.
문제는 무조건 1개의 정답만 있습니다.
문제를 풀이할 때 모든 선택지들을 검토하세요.
모든 선택지마다 근거를 지문에서 찾아 설명하세요.
추가지문이 있을 때, 반드시 추가지문을 참고하세요.
추가지문은 문제에서 <보기>로 표시될 수 있습니다.

다음의 형식을 따라 답변하세요.
정답: (정답)
1번: (선택지 1번에 대한 답변) + "(지문 속 근거가 된 문장)"
2번: (선택지 2번에 대한 답변) + "(지문 속 근거가 된 문장)"
3번: (선택지 3번에 대한 답변) + "(지문 속 근거가 된 문장)"
4번: (선택지 4번에 대한 답변) + "(지문 속 근거가 된 문장)"
5번: (선택지 5번에 대한 답변) + "(지문 속 근거가 된 문장)

"""
    prompt = f"""### 지문
{paragraph}

### 문제
{question}

### 추가지문
{question_plus}

### 선택지
{choices}

### 답변
"""
    #print(prompt)

    # Chat GPT
    completion = openai_client.chat.completions.create(
        model=gpt_model,
        messages=[
            {"role": "system", "content": base_prompt},
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    res = completion.choices[0].message.content
    res_answer = res.split('정답:')[-1].split('번')[0][-1].strip()
    ox = 'x'
    if(str(res_answer)==str(answer)):
        ox = 'o'

    rst = {
        'prompt':prompt,
        'gpt_completion': res,
        'gpt_answer': res_answer,
        'answer': answer,
        'ox': ox,
        'score': score
    }

    return rst
```

```python
import openai
client = openai.OpenAI(api_key=api_key)

rsts_4o = []
for idx, row in kice_df.iterrows():
    paragraph = row['paragraph']
    problems = row['problems']
    print(row['id'])
    for problem in problems:
        problem['paragraph'] = paragraph
        rst = prediction(problem=problem, openai_client=client, gpt_model='gpt-4o')

        rsts_4o.append(rst)
```

```python
rst_df = pd.DataFrame(rsts_4o)
rst_df
```

```python
total_score = rst_df[rst_df['ox']=='o']['score'].sum()

print('최종 점수:', total_score)
```
